```python
columns_to_drop=['DATETIME', 'CCN', 'OFFENSE_DISTRICT', 'OFFENSE_PSA',
       'DEFENDANT_PSA', 'DEFENDANT_DISTRICT','RACE', 'DESCRIPTION', 'ARREST_BLOCKX', 'ARREST_BLOCKY',
       'GIS_ID', 'CREATOR', 'CREATED', 'EDITOR', 'EDITED', 'OBJECTID',
       'GLOBALID']
marijuana_arrests_drop = df_marijuana_arrests.drop(columns=columns_to_drop, inplace=False)
print(marijuana_arrests_drop)
```

```python
marijuana_arrests_drop.head()
```

```python
df_ethnicity=pd.get_dummies(marijuana_arrests_drop['ETHNICITY'], drop_first=True)
df_ethnicity.drop(columns=['U','N'], inplace=True)
df_ethnicity = pd.get_dummies(df_ethnicity, drop_first=True, prefix='ETHNICITY')
df_ethnicity
#H=Hispanic or Latino
#N=Non-Hispanic or Latino
#U=Unknown
```

```python
df_sex= pd.get_dummies(marijuana_arrests_drop['SEX'])
df_sex.drop(columns=['U'],inplace=True)
print(df_sex)
#F: Female
#M: Male
#U: Unspecified/Unknown
```

```python
df_type=pd.get_dummies(marijuana_arrests_drop['TYPE'])
print(df_type.head())
#Type: Cultivation, Distribution, Manufacture, Possession, Possession with intent to distribute, Public consumption.
```

```python
df_category=pd.get_dummies(marijuana_arrests_drop['CATEGORY'])
print(df_category.head())

#1: Narcitics [Only one value exists]
```

```python
df_adult_juvenile=pd.get_dummies(marijuana_arrests_drop['ADULT_JUVENILE'])
print(df_adult_juvenile.head())
```

```python
df_adult=df_adult_juvenile.drop(columns=df_adult_juvenile[['Unknown','Juvenile']], inplace=False)
df_adult
```

```python
df_juvenile=df_adult_juvenile.drop(columns=df_adult_juvenile[['Unknown','Adult']], inplace=False)
df_juvenile
```

```python
df_main_marijuana_arrests=pd.concat([df_sex,df_ethnicity, df_type,df_adult,df_juvenile,marijuana_arrests_drop[['YEAR','AGE' ]]], axis=1)
df_main_marijuana_arrests
```

```python
sample_size=10000
df_sampled=df_main_marijuana_arrests.sample(n=sample_size, replace=False)
df_sampled
```

```python
df_sampled['AGE'].fillna(df_sampled['AGE'].mean(), inplace=True)
```

```python
X = df_sampled.drop(columns=['YEAR'])
y = df_sampled['YEAR']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
print(y.isnull().sum())
```

```python
decision_tree = DecisionTreeRegressor(random_state=4, max_depth=12)
decision_tree.fit(X_train, y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.tree import DecisionTreeRegressor, plot_tree
import matplotlib.pyplot as plt
from tabulate import tabulate

```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import seaborn as sns
```

```python
y_pred = decision_tree.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
table_data = [["Mean Squared Error (MSE)", mse],
              ["R-squared", r2]]
table = tabulate(table_data, headers=["Metric", "Value"], tablefmt="grid")
print(table)
```

```python
sns.set(style="whitegrid")
colors = sns.color_palette("viridis", n_colors=len(feature_importances_df))
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importances_df, palette=colors)
plt.xlabel('Importance', fontsize=12)
plt.ylabel('Feature', fontsize=12)
plt.title('Feature Importances', fontsize=16)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```
